```python
from diffusers import DiffusionPipeline
import torch
```

```python
model_id ="dreamlike-art/dreamlike-diffusion-1.0"
pipe1 = StableDiffusionPipeline.from_pretrained(model_id,torch_dtype=torch.float16)
pipe = pipe1.to("cuda")
```

```python
prompt = "a photo of a cute kitten sitting in the lap of a little girl"
images = pipe1(prompt).images[0]
plt.imshow(images)
plt.axis("off")
```

```python
md = "stabilityai/stable-diffusion-xl-base-1.0"
new = DiffusionPipeline.from_pretrained(md,torch_dtype=torch.float16,use_safetensors=True)
new = new.to("cuda")
```

```python
img = new(prompt).images[0]
plt.imshow(img)
plt.axis("off")
```
